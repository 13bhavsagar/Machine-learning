#1.What is a Decision Tree, and how does it work?
- A Decision Tree is a machine learning model that makes decisions by splitting data into branches based on questions about the features. It works like a flowchart, where each question leads to further splits until a final prediction is made. The tree chooses the best questions to ask using rules like information gain. It‚Äôs simple to understand and useful for both classification and regression tasks.


#2.What are impurity measures in Decision Trees?
- Gini Impurity ‚Äì Measures how often a randomly chosen element would be incorrectly labeled.

- Entropy ‚Äì Based on information theory, it measures the level of disorder or uncertainty in the dataset.

- Variance ‚Äì Used in regression trees to measure how much the target values vary within a node.

#5.What is Information Gain, and how is it used in Decision Trees?
- Information Gain is a measure used in Decision Trees to determine which feature gives the best split. It calculates how much the uncertainty (entropy) is reduced after the dataset is split based on a feature. The feature that provides the highest Information Gain is chosen for the split. This helps the tree grow in a way that best separates the data into distinct classes.

- $$
Information\ Gain = Entropy_{parent} - \sum_{k=1}^{n} \frac{N_k}{N} \cdot Entropy_k
$$

#6.What is the difference between Gini Impurity and Entropy?
- Gini Impurity and Entropy are both used in Decision Trees to measure how impure or mixed a node is, but they differ in calculation and interpretation. Gini Impurity measures the probability of incorrectly classifying a randomly chosen element, while Entropy measures the amount of uncertainty or disorder in the data. Gini is simpler and faster to compute as it doesn't use logarithms, whereas Entropy is based on information theory and uses log base 2. Although both often lead to similar splits, Gini is commonly used in the CART algorithm, while Entropy is used in ID3 and C4.5

#7.What is the mathematical explanation behind Decision Trees?
### Mathematical Explanation Behind Decision Trees

1.**Impurity Calculation**:

- **Gini Impurity**:
$$
Gini(D) = 1 - \sum_{i=1}^{n} p_i^2
$$

- **Entropy**:
$$
Entropy(D) = -\sum_{i=1}^{n} p_i \log_2 p_i
$$

Where \( p_i \) is the probability of class \( i \) in dataset \( D \), and \( n \) is the number of classes.

2.**Splitting the Dataset**:

Split the dataset \( D \) into subsets based on feature values.

3.**Information Gain or Gini Gain**:

To evaluate the quality of a split:
$$
Gain = Impurity_{parent} - \sum_{k=1}^{m} \frac{N_k}{N} \cdot Impurity_k
$$

Where:
- \( N \) is the total number of samples,
- \( N_k \) is the number of samples in subset \( k \),
- \( Impurity_k \) is the Gini or Entropy of subset \( k \),
- \( m \) is the number of resulting subsets.

4.**Best Split Selection**:

Choose the feature and threshold with the maximum gain (i.e., the highest impurity reduction).

5.**Recursive Tree Building**:

Repeat the process for each child node until a stopping condition is met (e.g., pure node, max depth).

#8.What is Pre-Pruning in Decision Trees?
- Pre-pruning is a technique used in Decision Trees to stop the tree from growing too large during its construction. It prevents splits that do not significantly improve model performance. This is done using conditions like maximum depth, minimum samples per node, or minimum information gain. Pre-pruning helps reduce overfitting and improves generalization.

#9.What is Post-Pruning in Decision Trees?
- Post-pruning is a technique used to simplify a fully grown Decision Tree by removing branches that have little importance. After the tree is built, it evaluates the performance of subtrees and prunes those that do not improve accuracy on validation data. This helps reduce overfitting by making the model less complex. Post-pruning improves the tree‚Äôs generalization to unseen data.

#10.What is the difference between Pre-Pruning and Post-Pruning?
- Pre-pruning and Post-pruning are techniques used to prevent overfitting in Decision Trees, but they differ in timing and approach. Pre-pruning stops the tree from growing too deep during its construction by applying limits such as maximum depth, minimum samples per node, or minimum information gain. In contrast, post-pruning allows the tree to grow fully and then removes unnecessary branches that do not improve performance, usually based on a validation set. While pre-pruning is faster and simpler, it might miss important patterns, whereas post-pruning tends to give better accuracy by simplifying the tree after analyzing its complete structure.

#11.What is a Decision Tree Regressor?
- A Decision Tree Regressor is a type of Decision Tree algorithm used for predicting continuous numerical values instead of categories. It works by splitting the dataset into regions based on feature values, just like in classification, but instead of predicting a class, it predicts the average value of the target variable in each region. At each node, the algorithm chooses the feature and split that minimizes the variance (or mean squared error) of the target values. It's simple, interpretable, and effective for capturing non-linear relationships in regression tasks.

#12.What are the advantages and disadvantages of Decision Trees?
**Advantages of Decision Trees**:
- Easy to understand and interpret ‚Äì resembles human decision-making.

-  No need for feature scaling ‚Äì works well with raw data.

- Handles both numerical and categorical data.

- Can model non-linear relationships effectively.

**Disadvantages of Decision Trees**
- Prone to overfitting, especially with deep trees.

- Unstable to small changes in data (can lead to a different tree).

- Biased with imbalanced datasets or dominant features.

- Less accurate than ensemble methods like Random Forests or Gradient Boosting.

#13.How does a Decision Tree handle missing values?
- A Decision Tree handles missing values by using methods like surrogate splits, assigning to the most frequent branch, or ignoring missing data during splits. In many cases, missing values are imputed before training.

#14.How does a Decision Tree handle categorical features?
- A Decision Tree handles categorical features by splitting them based on their unique values. For each categorical feature, the tree can create branches for each category (e.g., "Red", "Blue", "Green") or group categories based on impurity reduction. Most libraries, like scikit-learn, require categorical data to be encoded (e.g., one-hot or label encoding),

#23.Write a Python program to apply feature scaling before training a Decision Tree Classifier and compare its accuracy with unscaled data

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler

x,y=load_iris(return_X_y=True)
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
scaler=StandardScaler()
x_train_scaled=scaler.fit_transform(x_train)
x_test_scaled=scaler.transform(x_test)

clf1=DecisionTreeClassifier()
clf2=DecisionTreeClassifier()

clf1.fit(x_train,y_train)
clf2.fit(x_train_scaled,y_train)

y_pred1=clf1.predict(x_test)
y_pred=clf2.predict(x_test_scaled)

accuracy1=accuracy_score(y_test,y_pred1)
accuracy=accuracy_score(y_test,y_pred)

print(f"Accuracy of tree with unscaled data: {accuracy1}")
print(f"Accuracy of tree with scaled data: {accuracy}")
```

#27.Write a Python program to train a Decision Tree Classifier, apply Cost Complexity Pruning (CCP), and visualize its effect on accuracy

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

X,y = load_iris(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

clf = DecisionTreeClassifier(random_state=42)
clf.fit(X_train, y_train)

path = clf.cost_complexity_pruning_path(X_train, y_train)
ccp_alphas = path.ccp_alphas

clfs = []
for alpha in ccp_alphas:
    clf = DecisionTreeClassifier(random_state=42, ccp_alpha=alpha)
    clf.fit(X_train, y_train)
    clfs.append(clf)

train_scores = [clf.score(X_train, y_train) for clf in clfs]
test_scores = [clf.score(X_test, y_test) for clf in clfs]

plt.figure(figsize=(10, 6))
plt.plot(ccp_alphas, train_scores, marker='o', label='Train Accuracy', drawstyle="steps-post")
plt.plot(ccp_alphas, test_scores, marker='s', label='Test Accuracy', drawstyle="steps-post")
plt.xlabel("ccp_alpha (Complexity Parameter)")
plt.ylabel("Accuracy")
plt.title("Effect of Cost Complexity Pruning on Accuracy")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

#28.Write a Python program to train a Decision Tree Classifier and evaluate its performance using Precision, Recall, and F1-Score

```python
from sklearn.datasets import load_breast_cancer
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

X,y = load_breast_cancer(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

clf = DecisionTreeClassifier()
clf.fit(X_train, y_train)
y_pred=clf.predict(X_test)
print(f"Accuracy: {accuracy_score(y_test,y_pred)}")
print(f"Precision: {precision_score(y_test,y_pred,average='weighted')}")
print(f"Recall: {recall_score(y_test,y_pred,average='weighted')}")
print(f"F1-Score: {f1_score(y_test,y_pred,average='weighted')}")
```

#29.Write a Python program to train a Decision Tree Classifier and visualize the confusion matrix using seaborn

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_breast_cancer
from sklearn.metrics import confusion_matrix
x,y=load_breast_cancer(return_X_y=True)
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)
clf1= DecisionTreeClassifier(max_depth=5)
clf1.fit(x_train,y_train)
y_pred=clf1.predict(x_test)
con=confusion_matrix(y_test,y_pred)
sns.heatmap(con,annot=True,fmt='d')
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```
